```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**5

x_values = np.linspace(0, 1, 1000)

y_values = f(x_values)

# Tracé de la courbe
plt.plot(x_values, y_values, color = 'red', label='y = x⁵')

square_x = [0, 1, 1, 0, 0]
square_y = [0, 0, 1, 1, 0]

# Tracé du carré
plt.plot(square_x + [square_x[0]], square_y + [square_y[0]], color='purple', label='Square')


plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Courbe x⁵ et Carré délimité par x=0, x=1, y=0, y=1')
plt.legend()

# Affichage du graphe
plt.ylim(0, 1.4)  # Définir la limite y pour inclure y=1.4
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

class Curve:
    def __init__(self, start, stop, nbr_points=5432):
        self.start = start
        self.stop = stop
        self.nbr_points = nbr_points

    def _calculate_y(self, x):
        return x**5

    def plot_curve(self):
        x_values = np.linspace(self.start, self.stop, self.nbr_points)
        y_values = self._calculate_y(x_values)

        plt.plot(x_values, y_values, label='y = x⁵')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Courbe d\'équation x⁵ sur l\'intervalle [{}; {}]'.format(self.start, self.stop))
        plt.legend()

    def plot_points(self):
        x_points = np.linspace(self.start, self.stop, self.nbr_points)
        y_points = self._calculate_y(x_points)

        for x, y in zip(x_points, y_points):
            if y_points > self._calculate_y(x):
                plt.plot(x, y, 'b+', markersize=5)  # Point au-dessus de la courbe en bleu
            else:
                plt.plot(x, y, 'g.', markersize=5)  # Point en dessous de la courbe en vert

    def calculate_areas(self):
        x_values = np.linspace(self.start, self.stop, self.nbr_points)
        y_values = self._calculate_y(x_values)

        dx = (self.stop - self.start) / self.nbr_points
        area_blue = 0
        area_green = 0

        for x, y in zip(x_values, y_values):
            if y > self._calculate_y(x):
                area_blue += dx
            else:
                area_green += dx

        return area_blue, area_green

# Exemple d'utilisation de la classe
curve_instance = Curve(start=0, stop=1, nbr_points=1000)
curve_instance.plot_curve()
curve_instance.plot_points()
plt.show()

area_blue, area_green = curve_instance.calculate_areas()
print("Surface en bleu:", area_blue)
print("Surface en vert:", area_green)
```
